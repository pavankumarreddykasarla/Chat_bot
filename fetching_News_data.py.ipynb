```python
from datasets import load_dataset
import pandas as pd

# Load the dataset

dataset = load_dataset("dell-research-harvard/AmericanStories", "subset_years", year_list=[]);

# Define the batch size
batch_size = 10
start_year = 1744
end_year = 1900
```

```python
# Loop through the years in batches
for start in range(start_year, end_year, batch_size):
    end = min(start + batch_size, end_year)
    print(f"Processing data for years {start}-{end}")

    # Retrieve data for the current batch of years
    batch_data = dataset.filter(lambda example: start <= example["year"] < end)

    # Convert the batch data into a pandas DataFrame
    df = pd.DataFrame(batch_data)

    # Generate CSV file for the current batch
    csv_filename = f"data_{start}-{end}.csv"
    df.to_csv(csv_filename, index=False)

    print(f"CSV file '{csv_filename}' generated successfully.")
```
